```python
import numpy as np
import matplotlib.pyplot as plt

#   
amplitude = 1.0  #   
carrier_frequency = 10.0  #  
modulation_index = 0.5  #  

#   
t = np.linspace(0, 1, 1000)

#   
info_signal = amplitude * np.sin(2 * np.pi * t)

#   
carrier_signal = np.sin(2 * np.pi * carrier_frequency * t)

#    
modulated_signal = (1 + modulation_index * info_signal) * carrier_signal

#  
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, info_signal, 'b')
plt.title(' ')
plt.xlabel(', ')
plt.ylabel(', ^10-6')

plt.subplot(3, 1, 2)
plt.plot(t, carrier_signal, 'r')
plt.title(' ')
plt.xlabel(', ')
plt.ylabel(', ^10-6')

plt.subplot(3, 1, 3)
plt.plot(t, modulated_signal, 'g')
plt.title('  ')
plt.xlabel(', ')
plt.ylabel(', ^10-6')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#   
amplitude = 1.0  #   
carrier_frequency = 10.0  #  
modulation_index = 5.0  #  

#   
t = np.linspace(0, 1, 1000)

#   
info_signal = amplitude * np.sin(2 * np.pi * t)

#   
carrier_signal = np.sin(2 * np.pi * carrier_frequency * t)

#    
modulated_signal = np.sin(2 * np.pi * (carrier_frequency + modulation_index * info_signal) * t)

#  
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, info_signal, 'b')
plt.title(' ')
plt.xlabel(', ')
plt.ylabel(', ^10-6')

plt.subplot(3, 1, 2)
plt.plot(t, carrier_signal, 'r')
plt.title(' ')
plt.xlabel(', ')
plt.ylabel(', ^10-6')

plt.subplot(3, 1, 3)
plt.plot(t, modulated_signal, 'g')
plt.title('  ')
plt.xlabel(', ')
plt.ylabel(', ^10-6')

plt.tight_layout()
plt.show()
```
