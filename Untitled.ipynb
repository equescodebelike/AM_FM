```python
import numpy as np
import matplotlib.pyplot as plt

# Ввод значений амплитуды и частоты
amplitude = float(input("Введите амплитуду информационного сигнала: "))
carrier_frequency = float(input("Введите частоту несущей: "))

# Генерация временной оси
t = np.linspace(0, 1, 1000)

# Генерация информационного сигнала
info_signal = amplitude * np.sin(2 * np.pi * t)

# Генерация несущей сигнала
carrier_signal = np.sin(2 * np.pi * carrier_frequency * t)

# Генерация амплитудно модулированного сигнала
modulated_signal = (1 + info_signal) * carrier_signal

# Построение графиков
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, info_signal, 'b')
plt.title('Информационный сигнал')
plt.xlabel('Время, с')
plt.ylabel('Амплитуда, м^10-6')

plt.subplot(3, 1, 2)
plt.plot(t, carrier_signal, 'r')
plt.title('Несущий сигнал')
plt.xlabel('Время, с')
plt.ylabel('Амплитуда, м^10-6')

plt.subplot(3, 1, 3)
plt.plot(t, modulated_signal, 'g')
plt.title('Амплитудно модулированный сигнал')
plt.xlabel('Время, с')
plt.ylabel('Амплитуда, м^10-6')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Задание параметров модуляции
amplitude = float(input("Введите амплитуду информационного сигнала: "))
carrier_frequency = float(input("Введите частоту несущей: "))
modulation_index = 5.0  # Индекс модуляции

# Генерация временной оси
t = np.linspace(0, 1, 1000)

# Генерация информационного сигнала
info_signal = amplitude * np.sin(2 * np.pi * t)

# Генерация несущей сигнала
carrier_signal = np.sin(2 * np.pi * carrier_frequency * t)

# Генерация частотно модулированного сигнала
modulated_signal = np.sin(2 * np.pi * (carrier_frequency + modulation_index * info_signal) * t)

# Построение графиков
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, info_signal, 'b')
plt.title('Информационный сигнал')
plt.xlabel('Время, с')
plt.ylabel('Амплитуда, м^10-6')

plt.subplot(3, 1, 2)
plt.plot(t, carrier_signal, 'r')
plt.title('Несущий сигнал')
plt.xlabel('Время, с')
plt.ylabel('Амплитуда, м^10-6')

plt.subplot(3, 1, 3)
plt.plot(t, modulated_signal, 'g')
plt.title('Частотно модулированный сигнал')
plt.xlabel('Время, с')
plt.ylabel('Амплитуда, м^10-6')

plt.tight_layout()
plt.show()
```

## Амплитудная модуляция (АМ)
Амплитудная модуляция (АМ) - это метод модуляции, в котором изменение амплитуды информационного сигнала приводит к изменению амплитуды несущего сигнала, оставляя его частоту неизменной. Информационный сигнал кодируется в изменениях амплитуды несущего сигнала, что позволяет передавать информацию через канал связи. АМ широко используется в радио и телевизионном вещании, где аудиосигнал модулирует несущую радиоволну, чтобы передать звуковую информацию.

## Частотная модуляция (ЧМ)
Частотная модуляция (ЧМ) - это метод модуляции, в котором изменение частоты информационного сигнала приводит к изменению частоты несущего сигнала, оставляя его амплитуду неизменной. Информационный сигнал кодируется в изменениях частоты несущего сигнала, что позволяет передавать информацию через канал связи. ЧМ широко используется в радиосвязи, сотовой связи, спутниковой связи и других системах, где требуется высококачественная передача аудио или данных.
